```
import numpy as np
class NpEncoder(json.JSONEncoder):
    def default(self, obj):
        if isinstance(obj, np.integer):
            return int(obj)
        elif isinstance(obj, np.floating):
            return float(obj)
        elif isinstance(obj, np.ndarray):
            return obj.tolist()
        else:
            return super(NpEncoder, self).default(obj)
```

```
alias = {}
operator_dict = {'lt':1, 'eq':2, 'gt':3}
for i, query in enumerate(job_light['query']):
  fq = {}
  ands = []
  joins = []
  tbls = []
  print('working on ',i, 'th query')
  fq['Query Number'] = 'job_light_'+ str(i+1)
  fq['Query'] = query
  fq['Cardinality'] = job_light['cardinality_true'][i]
  tmp = parse(job_light['query'][i])
  #tmp = parse(query)
  #tables
  for tb in tmp['from']:
    pre = tb['value']
    apx = tb['name']
    tb_name = tables[pre]
    tb_alias = apx
    tbls.append(tb_name)
    alias[tb_alias] = pre
  for pred in tmp['where']['and']:
    op = list(pred.keys())
    if(type(pred[op[0]][1])==str):
      n_split = pred[op[0]][0]
      alias_split = n_split.split('.')
      act = alias[alias_split[0]]
      renamed_pred_1 = act+'.'+alias_split[1]
      #joins_tmp = []
      #joins_tmp.append(columns[renamed_pred])
      #joins_tmp.append(operator_dict[op[0]])
      n_split = pred[op[0]][1]
      alias_split = n_split.split('.')
      act = alias[alias_split[0]]
      renamed_pred_2 = act+'.'+alias_split[1]
      joins.append(str(columns[renamed_pred_1])+'-'+str(columns[renamed_pred_2]))
    elif(type(pred[op[0]][1])==int):
      ands_tmp = []
      n_split = pred[op[0]][0]
      alias_split = n_split.split('.')
      act = alias[alias_split[0]]
      renamed_pred = act+'.'+alias_split[1]
      ands_tmp.append(columns[renamed_pred])
      ands_tmp.append(operator_dict[op[0]])
      val = pred[op[0]][1]
      mIn = mins[columns[renamed_pred]]
      mAx = maxs[columns[renamed_pred]]
      if(mIn == None):
        mIn = 0
      if(mAx == None):
        mAx = 1
      norm_val = (val-mIn)/ (mAx-mIn)
      ands_tmp.append(norm_val)
      ands.append(ands_tmp)
  fq['AND Predicate Norm'] = ands
  fq['Tables'] = tbls
  fq['Joins'] = joins
  with open('JOB_light_'+str(i+1)+'.json', 'w', encoding='utf-8') as f:
    json.dump(fq, f, ensure_ascii=False, indent=4, cls=NpEncoder)
```

```
alias = {}
t = 0
i = 4000
for enum, query in enumerate(synthetic['query'][4000:5000]):
  fq = {}
  if(enum%100)==0:
    print('working on ',t+4000, 'th query')
    t = t + 100 
  fq['Query Number'] = 'Synthetic_'+ str(i+1)
  fq['Query'] = query
  fq['Cardinality'] = synthetic['cardinality_true'][i]
  tmp = parse(synthetic['query'][i])
  joins  = []
  tbls = []
  ands = []
  if(type(tmp['from'])==list):
    for tb in tmp['from']:
      pre = tb['value']
      apx = tb['name']
      tb_name = tables[pre]
      tb_alias = apx
      tbls.append(tb_name)
      alias[tb_alias] = pre
  elif(type(tmp['from'])==dict):
    pre = tmp['from']['value']
    apx = tmp['from']['name']
    tb_name = tables[pre]
    tb_alias = apx
    tbls.append(tb_name)
    alias[tb_alias] = pre
  a, b = list(tmp['where'].items())[0]
  if(a == 'and'):
    for pred in tmp['where']['and']:
      op = list(pred.keys())
      if(type(pred[op[0]][1])==str):
        n_split = pred[op[0]][0]
        alias_split = n_split.split('.')
        act = alias[alias_split[0]]
        renamed_pred_1 = act+'.'+alias_split[1]
        #joins_tmp = []
        #joins_tmp.append(columns[renamed_pred])
        #joins_tmp.append(operator_dict[op[0]])
        n_split = pred[op[0]][1]
        alias_split = n_split.split('.')
        act = alias[alias_split[0]]
        renamed_pred_2 = act+'.'+alias_split[1]
        joins.append(""+str(columns[renamed_pred_1])+'-'+str(columns[renamed_pred_2]))
      elif(type(pred[op[0]][1])==int):
        ands_tmp = []
        n_split = pred[op[0]][0]
        alias_split = n_split.split('.')
        act = alias[alias_split[0]]
        renamed_pred = act+'.'+alias_split[1]
        ands_tmp.append(columns[renamed_pred])
        ands_tmp.append(operator_dict[op[0]])
        val = pred[op[0]][1]
        mIn = mins[columns[renamed_pred]]
        mAx = maxs[columns[renamed_pred]]
        if(mIn == None):
          mIn = 0
        if(mAx == None):
          mAx = 1
        norm_val = (val-mIn)/ (mAx-mIn)
        ands_tmp.append(norm_val)
        ands.append(ands_tmp)
  elif(a =='eq' or a=='lt' or a=='gt'):
    op = a
    if(type(b[1])==str):
      n_split = b[0]
      alias_split = n_split.split('.')
      act = alias[alias_split[0]]
      renamed_pred_1 = act+'.'+alias_split[1]
      #joins_tmp = []
      #joins_tmp.append(columns[renamed_pred])
      #joins_tmp.append(operator_dict[op[0]])
      n_split = b[1]
      alias_split = n_split.split('.')
      act = alias[alias_split[0]]
      renamed_pred_2 = act+'.'+alias_split[1]
      joins.append(""+str(columns[renamed_pred_1])+'-'+str(columns[renamed_pred_2]))
    elif(type(b[1])==int):
      ands_tmp = []
      n_split = b[0]
      alias_split = n_split.split('.')
      act = alias[alias_split[0]]
      renamed_pred = act+'.'+alias_split[1]
      ands_tmp.append(columns[renamed_pred])
      ands_tmp.append(operator_dict[op])
      val = b[1]
      mIn = mins[columns[renamed_pred]]
      mAx = maxs[columns[renamed_pred]]
      if(mIn == None):
        mIn = 0
      if(mAx == None):
        mAx = 1
      norm_val = (val-mIn)/ (mAx-mIn)
      ands_tmp.append(norm_val)
      ands.append(ands_tmp)
  fq['AND Predicate Norm'] = ands
  fq['Tables'] = tbls
  fq['Joins'] = joins
  i = i+1
  with open('Synthetic_json/Synthetic_'+str(i)+'.json', 'w', encoding='utf-8') as f:
    json.dump(fq, f, ensure_ascii=False, indent=4, cls=NpEncoder)
  
```
